# Get coordinate for all the sprites

```
# Height and width for all the sprites in spritesheet
HEIGHT = 30 
WIDTH = 30

# Store ID, coordinate and dump to json file
sprite_coord = {} # ID: [x_upper_left, y_upper_left],[x_lower_right, y_lower_right]


# xy_upper_left = []
# xy_lower_right = []
# px_x, px_y = 0,0 
temp_coord = []

# ID = 0 contains HEIGHT pixel and WIDTH pixel of all the sprites that
# uses to create spritesheet
ID = 0
sprite_coord[0] = {'HEIGHT': 30, 'WIDTH': 30}

for px_y in range(0,img.shape[1],HEIGHT):
  for px_x in range(0,img.shape[0],WIDTH): 
    ID += 1
    sprite_coord[ID] = [px_x,px_y,px_x + WIDTH, px_y + HEIGHT]
# print(ID)
print(sprite_coord)
with open( 'crop_sprite_img/test.json', 'w') as file: 
  json.dump(sprite_coord, file)
```
